```python
# Import python packages as needed. Add your import statements in this block if you need more packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from sklearn.datasets import make_blobs
```

**Read the dataset into a numpy array**

```python
# Read the clustering dataset into a numpy array
df = pd.read_csv('../data/clustering_data.csv')
arr = df.to_numpy()
# Check if the data was read in correctly
print(arr)
```

**Plot the 2-D data to check if there are any groupings possible**

```python
# Add code here (you may add more cells as you need)
plt.scatter(arr[:, 0], arr[:, 1], color='cyan')
plt.title('Clustering Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()  

# there are 3 possible groupings
```

**Use the K-Means algorithm to group the datapoints that are similar**

```python
# Add code here (you may add more cells as you need)
#K-means chooses k cluster centers assigns point to nearest center updates center then repats until stable

from sklearn.cluster import KMeans


kmeans = KMeans(n_clusters = 3, random_state=42)
kmeans.fit(arr)

labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

**Plot the 2-D data color coded to the group assigned using the K-Means algorithm**

```python
# Add code here (you may add more cells as you need)
# data is assigned to cluster (0,1,2)
plt.scatter(arr[:, 0], arr[:, 1], c=labels)

# centroids are marked with red X
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200)

plt.title('K-means Clustering (Color-Coded by Cluster)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

**In a few lines describe why you chose the number of clusters n_clusters in the K-Means algorithm**

I chose 3 clusters because, as you can see from the scatter plot, there appears to be 3 similar sized sections, although the clusters are not perfectly uniform and have slight differences in spread and density.
